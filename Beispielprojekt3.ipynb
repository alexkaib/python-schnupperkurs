# Projekt 3: Wortverlaufsdiagramm

```python
import pandas as pd

search_word = "Gregor"

with open("verwandlung.txt", encoding="UTF-8") as file:
    text = file.read().replace("\n", " ")

words = text.replace(',', ' ').replace('.', ' ').split(' ')
occurance_positions = []
word_count = 0

for word in words:
    word_count += 1
    if word.lower() == search_word.lower():
        occurance_positions.append(word_count)

pd.Series(occurance_positions).plot(kind="hist")
```
