# Projekt 2: Häufigkeitsvergleich

```python
import pandas as pd

search_words = {
    "Gregor": 0,
    "Vater": 0,
    "Mutter": 0,
    "Schwester": 0
}

# zähle Anzahl von Suchbegriffen
with open("verwandlung.txt", encoding="UTF-8") as file:
    text = file.read().replace("\n", " ")

words = text.replace(',', ' ').replace('.', ' ').split(' ')

for word in words:
    if word in search_words:
        search_words[word] += 1

pd.Series(search_words).plot(kind="pie")
```
