# Teil 3: Bedingungen
**Bedingungen** modellieren Entscheidungen in einem Programm. Je nachdem, ob ein bestimmter Ausdruck **wahr** oder **falsch** ist, wird ein bestimmter Abschnitt des Programms (ein **Code-Block**) ausgef√ºhrt oder √ºbersprungen. Dadurch k√∂nnen **Verzweigungen** in ein Programm eingebaut werden, so dass auf verschiedene Situationen unterschiedlich reagiert werden kann.

## 3.1 Vergleichsoperatoren

Die einfachsten Aussagen, anhand derer eine Entscheidung getroffen werden kann, sind **Vergleiche**. Stimmt eine Nutzereingabe mit einem Password √ºberein? Ist der Wert eines Gutscheins h√∂her als der Einkaufspreis? Ist es vor oder nach 17:00 Uhr? Das alles sind Fragen, die in Python mit Vergleichsoperatoren beantwortet werden k√∂nnen, um einen **Wahrheitswert** zu erhalten.

### (Un-)Gleichheit (`==` und `!=`) 

````python
# Der == Operator: Sind zwei Werte gleich?

secret_password = "swordfish"
user_password = input("Was ist das Passwort? ")

secret_password == user_password
````

````python
# Der != Operator: Sind zwei Werte ungleich?

my_name = "Alex"
your_name = input("Wie hei√üt du? ")

my_name != your_name
````

### Gr√∂√üer und kleiner (`>` und `<`)

````python
# Der > Operator: Ist links gr√∂√üer als rechts?

gift_card = 10.0
book_price = 7.5

gift_card > book_price
````

````python
# Der < Operator: Ist links kleiner als rechts?

current_time = 13 
quitting_time = 17

current_time < quitting_time
````

### Gr√∂√üer-/Kleiner-gleich (`>=` und `<=`)

````python
# Der >= Operator: Ist links gr√∂√üer oder gleich rechts?

2 >= 2.0
````

````python
# Der <= Operator: Ist links kleiner oder gleich rechts?

4 <= 5
````

### Enthalten-sein (`in`)

Mit dem Schl√ºsselwort `in` kann √ºberpr√ºft werden, ob ein String in einem anderen String enthalten ist. Streng genommen ist das keine Vergleichsoperation, aber es wird ebenfalls ein Wahrheitswert produziert.

````python
# Ist das Wort im Satz enthalten?
sentence = "Lass uns einen Film schauen."
search_word = "Film"

search_word in sentence
````

## 3.2 Wenn-dann

Die grundlegende **Bedingungsanweisung** (wenn x, dann y) besteht aus dem Schl√ºsselwort `if`, einem als wahr oder falsch bewerteten `Ausdruck` und einem Doppelpunkt `:`. 

Danach folgt **einger√ºckt** (mit der Tabulatortaste `‚Üπ`, links √ºber Umschalt; oder mehreren Leerzeichen) ein Code-Block, der nur ausgef√ºhrt wird, wenn der Ausdruck wahr ist. 

Der Ausdruck nach `if`, dessen Wahrheitswert bestimmt wird, ist die eigentliche **Bedingung**. Der einger√ºckte Code-Block ist die **bedingte Anweisung**.

````python
is_raining = True

# Die Bedingung: Wenn es regnet...
if is_raining:
    # Die bedingte Anweisung: ... dann erinnere mich an den Regenschirm.
    print("Vergiss nicht deinen Regenschirm.")
````

````python
# Noch ein Beispiel

if 1 + 1 == 2:
    print("Alles gut, Mathe funktioniert noch.")
````

## 3.3 Mehrzeilige Bl√∂cke

**Bl√∂cke** k√∂nnen mehrere Zeilen umfassen, solange die Zeilen mit dem **gleichen Abstand** einger√ºckt sind.

````python
# Bedingung mit mehreren Zeilen
number = 1

if number < 5:
    print("Die Zahl ist zu klein! Wir machen sie gr√∂√üer...")
    number = number + 10
    print("Jetzt ist die Zahl: " + str(number))
````

## 3.4 Verschachtelte Bl√∂cke

Einen Block in einem anderen Block nennt man **verschachtelt**. So lassen sich zum Beispiel Bedingungen umsetzen, die nur dann gepr√ºft werden, wenn eine vorherige Bedingung erf√ºllt ist.

Hier ist es besonders wichtig, auf die **Einr√ºckungen** zu achten, da sich sonst leicht Fehler einschleichen k√∂nnen!

````python
available_stock = 20
discount = True

if available_stock > 0:
    print("Gehe zum Einkauf √ºber...")
    if discount:
        print("Sie erhalten auf den Einkauf Rabatt!")
````

## 3.5 Wenn-dann-sonst

Oft m√∂chte man mit einer Bedingung nicht nur auf die Wahrheit einer Aussage reagieren, sondern auch bei ihrer Falschheit etwas bestimmtes tun.

Das wird durch das `if... else...` Konstrukt erreicht.

````python
available_money = 12
total_price = 20

# Eine normale Bedingung...
if available_money >= total_price:
    print("Danke f√ºr Ihren Einkauf")

# Die "else"-Anweisung: der darauffolgende Codeblock wird nur ausgef√ºhrt, wenn die vorherige Bedingung falsch ist
else:
    print("Fehler: Guthaben reicht nicht aus.") 
````

# Teil 5: For-Schleifen
**Schleifen** f√ºhren einen **Code-Block** mehrmals aus. Die `for`-Schleife geht dabei jedes Element einer **Liste** (o.√§. **Datenstrukturen**) durch.

## 5.1 Listen iterieren

Die grundlegende Syntax f√ºr `for`-Schleifen in Python ist folgende:
```python
for element in list_name:
    ...
```
Der einger√ºckte Code, hier mit `...` angedeutet, ist der **Schleifenk√∂rper**. Er wird so oft wiederholt, wie es Elemente in der Liste gibt. Au√üerdem wird die hier `element` genannte Variable bei jedem Durchgang mit dem n√§chsten Element in der Liste √ºberschrieben. So kann man z.B. jedes Element bearbeiten oder nach bestimmten Elementen suchen.

Die Elemente einer Liste durchzugehen wird auch als das **Iterieren** der Liste bezeichnet. Jeder Durchgang ist dabei eine **Iteration**.

````python
words = ["This", "is", "a", "sentence"]

for word in words:
    print(word)
````

Der Name nach `for` ist eine Variable und kann beliebig gew√§hlt werden. Der folgende Code erreicht beispielsweise genau dasselbe wie der vorhergehende:

````python
words = ["This", "is", "a", "sentence"]

# Hier wird der Variablenname "thing" statt "word" benutzt - das macht aber keinen Unterschied
for thing in words:
    print(thing)
````

## 5.2 Iterationen √ºberspringen

Manchmal m√∂chte man eine Iteration, also einen Durchgang der Schleife, √ºberspringen. Daf√ºr wird der `continue` Befehl benutzt, der den Computer dazu veranlasst, sofort die n√§chte Iteration zu beginnen.

````python
# Beispiel: Ein- und Ausz√ºge auf einem Bankkonto
transactions = [300, -9.99, -24, -1.99, 160, -210, -290, 500]
revenue = 0

for transaction in transactions:
    # Negative Transaktionen werden √ºbersprungen, um nur die Einnahmen zu addieren
    if transaction < 0:
        continue
    revenue = revenue + transaction

print(revenue)
````

### üêûBug Hunt: E-Mail-Validierung
Was stimmt an dem folgenden Code nicht? Analysiere das Ergebnis und versuche, den Code zu verbessern!

````python
emails = ['user@example.com', 'invalid-email', 'test@domain.com', '', 'hello@world.net']
valid_emails = []

for email in emails:
    if '@' not in email:
        continue
        valid_emails.append(email)
    
print("G√ºltige Email-Adressen:", valid_emails)
````

## 5.3 Schleifen unterbrechen

Manchmal m√∂chte man nicht nur eine Iteration, sondern **alle Iterationen** √ºberspringen. Daf√ºr wird der `break` Befehl benutzt, der den Computer dazu veranlasst, die Schleife sofort zu beenden.

````python
transactions = [300, -9.99, -24, -1.99, 160, -210, -290, 500]
balance = 0

balance = 0
for transaction in transactions:
    balance = balance + transaction
    if balance < 0:
        print("Achtung: Konto√ºberziehung!")
        break
````
