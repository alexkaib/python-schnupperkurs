# ✏️ Musterlösung: Textvergleich

## Aufgabenstellung

Führe die folgende Code-Zelle aus, um die Variablen `verwandlung` und `zauberberg` zu erhalten. Anschließend kannst du auf sie zurückgreifen, um die untenstehende Aufgabe zu lösen.

```python
# Diese Code-Zelle ausführen, um Zugriff auf die beiden Texte als Variablen zu erhalten
with open("verwandlung.txt", encoding="UTF-8") as file:
    verwandlung = file.read().replace("\n", " ")

with open("zauberberg.txt", encoding="UTF-8") as file:
    zauberberg = file.read().replace("\n", " ")
```

In den Variablen `verwandlung` und `zauberberg` (Achtung Kleinschreibung!) sollten nun die Volltexte der beiden Werke von Franz Kafka und Thomas Mann als String enthalten sein. Führe einen quantitativen Stilvergleich aus, indem du folgende Fragen beantwortest:
- Welches Werk hat durchschnittlich längere Sätze (Wörter pro Satz)?
- Welches Werk hat durchschnittlich längere Wörter?
- Welches Werk besitzt anteilig mehr Fragesätze?
- Was ist das längste Wort in beiden Texten?

**Hinweis**: Es gibt verschiedene Möglichkeiten, diese Fragen mit Python zu beantworten, aber für die meisten werden **String-Methoden** benötigt. Lese dir folgende Übersicht durch und überlege, welche hilfreich sein könnten: https://www.w3schools.com/python/python_ref_string.asp

## Lösungsvorschlag
Hinweis: Der folgende Code liefert mit einfachen Mitteln eher Annhäherungen auf die Fragen; methodisch exakt ist das Vorgehen nicht.

```python
# Welches Werk hat durchschnittlich längere Sätze (Wörter pro Satz)?

sentences_v = verwandlung.count(".") + verwandlung.count("?") + verwandlung.count("!")
words_v = len(verwandlung.split())

print("Verwandlung: " + str(words_v / sentences_v) + " Wörter pro Satz")

sentences_z = zauberberg.count(".") + zauberberg.count("?") + zauberberg.count("!")
words_z = len(zauberberg.split())

print("Zauberberg: " + str(words_z / sentences_z) + " Wörter pro Satz")
```

```python
# Welches Werk hat durchschnittlich längere Wörter?

characters_v = len(verwandlung.replace(" ", ""))
words_v = len(verwandlung.split())

print("Verwandlung: " + str(characters_v / words_v) + " Zeichen pro Wort")

characters_z = len(zauberberg.replace(" ", ""))
words_z =len(zauberberg.split())

print("Zauberberg: " + str(characters_z / words_z) + " Zeichen pro Wort")
```

```python
# Welches Werk besitzt anteilig mehr Fragesätze?

questions_v = verwandlung.count("?")
sentences_v = verwandlung.count(".") + verwandlung.count("?") + verwandlung.count("!")

print("Verwandlung: " + str(100 * questions_v / sentences_v) + " Prozent Fragesätze")

questions_z = zauberberg.count("?")
sentences_z = zauberberg.count(".") + zauberberg.count("?") + zauberberg.count("!")

print("Zauberberg: " + str(100 * questions_z / sentences_z) + " Prozent Fragesätze")
```

```python
# Was ist das längste Wort in beiden Texten?

words_v = verwandlung.replace('.', ' ').replace('?', ' ').replace('!', ' ').replace(',', ' ').replace(';', ' ').split(' ')
longest_word_v = ""
for word in words_v:
    if len(word) > len(longest_word_v):
        longest_word_v = word

print("Längstes Wort Verwandlung: " + longest_word_v)

words_z = zauberberg.replace('.', ' ').replace('?', ' ').replace('!', ' ').replace(',', ' ').replace(';', ' ').split(' ')
longest_word_z = ""
for word in words_z:
    if len(word) > len(longest_word_z):
        longest_word_z = word

print("Längstes Wort Zauberberg: " + longest_word_z)
```
