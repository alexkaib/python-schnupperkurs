# Projekt 1: Worthäufigkeit

```python
search_word = "Mutter"
amount_first_half = 0
amount_second_half = 0

# zähle Anzahl der Zeilen
total_lines = 0
with open("verwandlung.txt", encoding="UTF-8") as file:
    for line in file:
        total_lines += 1

# zähle Anzahl von Suchbegriff in erster und zweiter Hälfte
with open("verwandlung.txt", encoding="UTF-8") as file:
    current_line = 0
    for line in file:
        current_line += 1
        
        words = line.split(" ")
        for word in words:
            if word.lower() == search_word.lower():
                if current_line / total_lines <= 0.5:
                    amount_first_half += 1
                else:
                    amount_second_half += 1

print("Das Wort '" + search_word + "' taucht " + str(amount_first_half + amount_second_half) + " mal auf.")
print("Erste Hälfte:", amount_first_half)
print("Zweite Hälfte:", amount_second_half)
```
