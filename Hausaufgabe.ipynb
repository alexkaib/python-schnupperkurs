### Tipp 4

Der Anteil an Fragesätzen ist die Anzahl an Sätzen, die mit einem Fragezeichen enden, geteilt durch die Gesamtzahl an Sätzen (also alles, was auf Punkt, Frage- oder Ausrufezeichen endet).

### Tipp 5

### Tipp 6

Um das längste Wort in einem Text zu erhalten, solltest du zunächst eine Variable (z.B. `longest_word`) mit dem leeren String `""` als Inhalt einrichten. Gehe dann jedes Wort im Text durch und prüfe seine Länge. Wenn es länger als der Inhalt von `longest_word` ist, überschreibe die Variable mit diesem Wort. Das Wort, das am Ende der Schleife in `longest_word` steht, ist das längste (oder genauer: es gibt im Text kein längeres Wort).
